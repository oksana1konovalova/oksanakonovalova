```python
next_user_stats = np.array([0, 1, 2, 0, 0, 0])
```

```python
for i in enumerate (users_stats):
    print (i)
```

```python
iLength = np.sqrt( (users_stats*users_stats).sum() )
next_user_statsLength = np.sqrt( (next_user_stats*next_user_stats).sum() )
```

```python
def cosine( i, next_user_stats ):
    iLength = np.linalg.norm( i )
    next_user_statsLength = np.linalg.norm( next_user_stats )
    return np.dot( i, next_user_stats ) / ( iLength * next_user_statsLength )
```

```python
np.arccos(cosine( users_stats, next_user_stats ))
```
