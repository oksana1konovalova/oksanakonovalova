```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = pd.read_csv('googleplaystore.csv')
names.head(50)
```

```python
names_barh = names[names['Content Rating']=='Teen'].sort_values('Installs', ascending=False).head(15)
names_barh
```

```python
plt.barh(names_barh['App'], names_barh['Rating']) 
```

```python
plt.barh(names_barh['App'], names_barh['Rating'])
plt.xlabel('Рейтинг')
plt.ylabel('Приложение')
plt.title('Топ приложений Google Play')
plt.text(5.0, 3.0, 'Самый низкий рейтинг')
plt.grid(15)
```

```python
names_barh_d = names_barh.groupby('Size').mean()
```

```python
names_barh_d.plot()
```

```python
# Видно, что размер приложения не зависит от рейтинга. Например, приложение с самым высоким рейтингом не имеет самый большой размер.
```

```python
names_barh_dynamics = names_barh.groupby([names_barh.Genres,'Genres']).sum()
names_barhn_dynamics = names_barh_dynamics.loc[(names_barh_dynamics.index.get_level_values(0).isin(['Social', 'Communication']))]
names_barh_dynamics.unstack().plot(kind='bar')
```

```python
#Видно, что сумма рейтингов приложений Google Play жанра Общение больше, чем сумма рейтингов приложений других жанров  
```

```python
names_dynamics = names.groupby(['Content Rating']).sum().sort_values(by='Rating', ascending=False).head(2)
names_dynamics.plot(kind='pie', y='Rating')
```

```python
#Количество рейтингов приложений для всех пользователей больше количества рейтингов приложений только для подростков
```
